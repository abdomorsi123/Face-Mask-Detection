```python
from flask import Flask,jsonify,request,send_file
from face_mask import MaskPrediction
mp = MaskPrediction()
```

```python

app = Flask(__name__) ## __name__= current file name  (main)


@app.route("/") ## page name 
def hello_world():
    result = {"status":"Success","Application":"Face Mask Detection"}
    return jsonify(result)

@app.route("/GetData") ## page name 
def get_data():
    mp.get_data()
    result = {"status":"data obtained successfully","Application":"Mask Detection"}
    return jsonify(result)

@app.route("/preprocess") ## page name 
def preprocess():
    mp.preprocess()
    result = {"status":"data preprocessed successfully","Application":"Mask Detection"}
    return jsonify(result)

@app.route("/classify") ## page name 
def classify():
    acc=mp.classify()
    result = {"status":"data classified successfully","Application":"Mask Detection","accuracy":acc}
    return jsonify(result)

@app.route("/predict") ## page name 
def predict():
    image = request.args['image']
    image1= mp.predict(image)
    return send_file(f'C:/Users/Abdalrhman Morsi/Desktop/API face mask/{image1}.jpeg', mimetype='image/jpeg', attachment_filename=f'{image1}.jpeg')


app.run()
```
